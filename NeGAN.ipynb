```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import save_img
from PIL import Image
import PIL.ImageOps
import os
```

```python
in_directory = "/content/drive/My Drive/Faces"
out_directory ="/content/drive/My Drive/post_photos"

onlyfiles = [f for f in os.listdir(in_directory)]
```

```python
j=1
for i in onlyfiles:
  try:
    img = load_img(f"{in_directory}/{i}")
    inverted_image = PIL.ImageOps.invert(img)
    save_img(f"{out_directory}/{i}",inverted_image)
    print(j)
    j=j+1
  except:
    pass
```
